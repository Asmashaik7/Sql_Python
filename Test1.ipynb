```python
# 8. Given two sets:
#  A = {1, 2, 3, 4} 			B = {3, 4, 5, 6}
#  Write code to print:
# Union
# Intersection
# Difference (A − B)
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A.union(B))
```

```python
# Union-all elements from both
print(A|B)# pipe symbol-using union function
```

```python
# Intersection-common from both
print(A & B)
```

```python
print(A.intersection(B)) # using intersection function
```

```python
# Difference (A − B)-A but not B
print(A-B)
```

```python
#using difference function
A.difference(B)
```

```python
#symmentric difference-A and B, but not in both
A.symmetric_difference(B)
```
