# **Problem Statement:**

## We want to analyze the sales amount of products across different regions to understand the distribution, central tendency, and variability of sales. This can help the business identify regions with high sales, consistency, or outliers in their sales data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample sales data (Region, Product Category, Sales Amount)
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Product_Category': ['Electronics', 'Clothing', 'Groceries', 'Electronics', 'Clothing', 'Groceries', 'Electronics', 'Clothing', 'Groceries', 'Electronics'],
    'Sales_Amount': [1200, 800, 900, 1500, 700, 950, 1300, 850, 1000, 1600]
}
```

```python
data
```

```python
# Create a Dataframe
df=pd.DataFrame(data)
```

```python
df
```

 ## Calculate Measures of Central Tendency

```python
mean_sales = df['Sales_Amount'].mean()
median_sales = df['Sales_Amount'].median()
mode_sales = df['Sales_Amount'].mode()[0]
```

```python
# In this dataset, all values occur only once, so technically the dataset does not have a statistical mode.
# However, for demonstration purposes and to maintain consistency with the instructorâ€™s example, we use mode()[0], which returns the first value (700).
# This approach is commonly used in practice when teaching central tendency measures.
```

```python
# Display the results
print(f"Mean Sales Amount: ", mean_sales)
print(f"Median Sales Amount: {median_sales}")
print(f"Mode Sales Amount: {mode_sales}")
```

## Calculate Measures of Dispersion

```python
variance_sales = df['Sales_Amount'].var()

sd_sales = df['Sales_Amount'].std()
```

```python
print("\nMeasures of Dispersion:")
print(f"Variance of Sales Amount: {round(variance_sales, 2)}")
print(f"Standard Deviation of Sales Amount: {round(sd_sales, 2)}")
```

```python
# Range is a measure of dispersion that shows the difference between the highest and lowest values.
# In Python (pandas), we calculate it using:
# max() - min().
range_sales=df['Sales_Amount'].max()-df['Sales_Amount'].min()
print(f"Range :{range_sales}")
```

## Calculate Quartiles and percentiles

```python
# These are findout by using Numerical function from NUmpy.
quartiles= np.percentile(df['Sales_Amount'],[25,50,75])
percentiles = np.percentile(df['Sales_Amount'], [10, 90])
```

```python
print(f"quartiles : {quartiles}")
print(f"percentiles : {percentiles}")
```

```python
print("\nQuartiles and Percentiles:")
print(f"25th Percentile (Q1): {quartiles[0]}")
print(f"50th Percentile (Median): {quartiles[1]}")
print(f"75th Percentile (Q3): {quartiles[2]}")
print(f"10th Percentile: {percentiles[0]}")
print(f"90th Percentile: {percentiles[1]}")
```

# Visualization of Sales Distribution

```python
plt.figure(figsize=(6, 4))
plt.hist(df['Sales_Amount'], bins=5, edgecolor='black', alpha=0.7)
plt.axvline(mean_sales, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_sales}')
plt.axvline(median_sales, color='blue', linestyle='dashed', linewidth=1, label=f'Median: {median_sales}')
plt.axvline(mode_sales, color='green', linestyle='dashed', linewidth=1, label=f'Mode: {mode_sales}')
plt.title('Sales Amount Distribution')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
