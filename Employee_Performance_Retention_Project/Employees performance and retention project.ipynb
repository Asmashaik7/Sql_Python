## 2. Data Cleaning and Preprocessing

### Handling Missing Values

```python
df.isna().sum()
```

```python
df['education'].mode() # to fill the missing eductaion with mode-here its bachelors
```

```python
df['education'] = df['education'].fillna(df['education'].mode()[0])
```

```python
df.info()
```

```python
# handle previous year rating
df['previous_year_rating'] = df['previous_year_rating'].fillna(
    df['previous_year_rating'].median()
)
```

```python
df.info()
```

```python
df.isna().sum()
```

### Missing values in the education and previous_year_rating columns were handled appropriately. The education column was imputed using the mode, while previous_year_rating was imputed using the median to maintain robustness against outliers. After imputation, the dataset contained no missing values and was ready for exploratory analysis.

# Step 4: Recommendations & Conclusion

## 4. Recommendations

Based on the exploratory data analysis, the following recommendations are proposed to improve employee productivity and reduce turnover:

1. **Strengthen Targeted Training Programs**  
   Employees with higher average training scores consistently met KPIs and demonstrated longer tenure. The organization should focus on improving training quality and tailoring programs by department to enhance performance outcomes.

2. **Focus on Early-Career Employee Retention**  
   Employees with shorter tenure and lower age groups exhibited higher attrition risk. Structured onboarding programs, mentorship initiatives, and early performance support can help improve retention among early-career employees.

3. **Leverage Performance Ratings for Retention Strategy**  
   Employees with higher previous-year ratings tended to stay longer. Regular performance feedback, transparent appraisal systems, and growth-oriented discussions can improve engagement and retention.

4. **Expand Recognition and Rewards Programs**  
   Award-winning employees showed higher average tenure, indicating that recognition positively impacts retention. Expanding reward and recognition initiatives can motivate employees and reduce turnover.

5. **Department-Specific Interventions**  
   Certain departments demonstrated lower average tenure and performance. Customized engagement strategies, leadership support, and department-level training interventions should be implemented to address these gaps.


## 5. Conclusion

This HR analytics project analyzed employee performance and retention-related trends using demographic, training, and performance data. The findings highlight a strong relationship between training effectiveness, performance outcomes, and employee tenure. Employees who achieved KPIs, received awards, and demonstrated strong training performance tended to remain longer within the organization.

The analysis also identified early-career employees and specific departments as higher retention risk areas, emphasizing the need for targeted engagement and support strategies. Overall, the insights derived from this analysis can help HR teams design data-driven interventions to improve employee productivity, engagement, and long-term retention.
