## Creating, Saving, and Loading a Pandas DataFrame in Multiple Formats (CSV, JSON, Excel)

```python
# Pandas is format-independent — as long as data can be structured like a table, Pandas can read it.
# Create DataFrame manually from a dictionary
# Save DataFrame to CSV, JSON, Excel
# Load DataFrame from CSV, JSON, Excel
```

# Step 1: Imorting pandas

```python
import pandas as pd
```

# Step 2: Creating a Dictionary

```python
# Create DataFrame manually from a dictionary
data={
    'Name': ['Asma','Manha','Ayman','Jaffer'],
    'Age': [38,6,10,40],
    'City': ['Kgm','Hyd','NK Nagar','Mancherial']
}
data
```

```python
# column names as keys (Name, Age, City)
# rows as list values. df are table like structure, 2D
# This is like creating a table manually inside Python.
#convert this dictionary into df
```

# Step 3: Convert Dictionary → DataFrame

```python
df=pd.DataFrame(data)
df
# Pandas directly converts dictionary → table.
```

```python
# Now df is a DataFrame (2D table)
```

## Step 4: Saving a DataFrame into different formats

## df into csv format

```python
df.to_csv("people.csv",index=False)

# Saves the table as a CSV file.
# index=False means don't add row numbers in the file.
```

```python
#df.to_excel("people.xlsx", index=False)
```

```python
df.to_json("people.json", orient="records")
# Saves the data as JSON format.
# orient="records" means store each row as a separate JSON object.
```

## Step 5: Load These Files Back Into Pandas

```python
df_csv=pd.read_csv("people.csv")
df_json=pd.read_json("people.json")
#df_excel=pd.read_excel("people.xlsx")
```

```python
#Now Pandas reads each file type and converts it back into DataFrames, we gave new names to df.
```
