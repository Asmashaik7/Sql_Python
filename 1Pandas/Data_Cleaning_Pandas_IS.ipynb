```python
# Creating a copy of dataframes
```

```python
import pandas as pd
```

```python
orders=pd.read_csv(r'D:\Internshala BI\M5-20251101T095706Z-1-001\M5\orders.csv')
```

```python
orders.head(3)
```

```python
# Craeting the copy of the dataframes
```

```python
orders_df=orders
orders_df
```

```python
# here top 3 rows are empty and some NaN values are also there in other rows. we can remove Nan values or can impute in them.
```

```python
# Remove top 3 empty rows
orders_df=orders_df.iloc[3:].reset_index(drop=True) # 
```

```python
orders_df.head(5)
```

```python
# Replace incorrect values( Fix date issues)
orders_df['Order_Date']=orders_df['Order_Date'].replace('11/01/2025','11-01-2025')
orders_df
```

```python
orders_df.info()
```

```python
# Step3: Remove rows with Null values in critical columns.Here we can observe from the above info(), have null in Quantity, Order_Date and Total_Amount.
```

```python
orders_df.dropna(subset=['Quantity', 'Order_Date','Total_Amount'], inplace=True)
orders_df
```

```python
#orders_df.dropna.We can use till here also inorder to remove null values but by using subset we tell pandas to remove particularly from those columns
#inplace=True means,normally without inplace pandas will show the df without null values temporarily. 
#We want to remove null values permanantly from orders_df
```

```python
# Change the Datatype of (Order_Date  13 non-null  object ), Its object given here, Change into DateTime. coerce is handle errors
```

```python
orders_df['Order_Date']=pd.to_datetime(orders_df['Order_Date'],errors='coerce')
orders_df
```

```python
# Convert Quantity column into Integer( quantity -float seems bad )
# astype to change data type, here into integer.
orders_df['Quantity']=orders_df['Quantity'].astype(int)
orders_df
```

```python
orders_df.info() #chcek the status of changes
```

```python
# Step 5: Removing duplicate rows
orders_df.drop_duplicates(inplace=True)
orders_df
```

```python
# Step 6: Adding a calculated column (unit price)
orders_df['Unit Price']=orders_df['Total_Amount']/orders_df['Quantity']
orders_df
```

```python
# Step 7: Remove unwanted columns(Extra_column)
orders_df.drop(columns=['Extra_Column'], inplace=True)
orders_df
```
